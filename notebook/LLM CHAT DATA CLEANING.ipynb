```python
def convert_to_conversation(df):
    conversation = ''
    for index, row in df.iterrows():
        text = ast.literal_eval(row[0])
        if 'Bot' in text[0]['text']['content'] or 'bot' in text[0]['text']['content']:
            text_content = re.sub(r'<b>|</b>|<p>|</p>|<br>', '', text[0]['text']['content'])
            text_content = clean(text_content, no_emoji=True)
            #text_content = text_content.replace('\n\n', '').replace('\r', '')
            text_content = text_content.replace('Bot', '### Assistant:')
            text_content = text_content.replace('Customer', '### Human:')
            text_content = text_content.replace('customer', '### Human:')
            text_content = text_content.replace('bot', '### Assistant:')
            
            start_phrase = '###agent: please describe the issue you are experiencing'
            end_phrase = '###agent: connecting you to our customer care representative.'
            start_index = text_content.find(start_phrase)
            end_index = text_content.find(end_phrase)
            #print(text_content)
            

            if start_index != -1 and end_index != -1:
                conversation += text_content[start_index:end_index].strip()
                #print('conversation')
        
        else:
            actor_type = "###" + row['actor_type']
            content = text[0]['text']['content']
            content = clean(content, no_emoji=True)
            conversation += actor_type + ": " + content + " "
    return conversation.strip()
```

```python
conversation_df = pd.DataFrame(columns=['conversation'])
```

```python
chat_id = list(final['interaction_id'].unique())
conversation=[]
for i in chat_id:
    try:
        df = final[final['interaction_id']==i][['message_parts','actor_type','message_source']]
        df = df[df.message_source != 'system']
        data = convert_to_conversation(df)
        conversation.append(data)
    except KeyError:
        None 
df = pd.DataFrame(conversation, columns=['conversation'])
conversation_df = pd.concat([df, conversation_df])
```
